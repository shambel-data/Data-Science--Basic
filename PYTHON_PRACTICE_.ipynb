```python
import pandas as pd
df=pd.read_csv('/content/sample_data/california_housing_train.csv')
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
result=df[(df['median_house_value']>500000) &(df['total_rooms']>5)]
print(len(result))
```

```python
Rich=df[df['median_house_value']>500000]
mean_median_income = Rich['median_income'].mean()
print(round(mean_median_income, 2))
```

```python
medium=df[df['median_house_value']<=500000]
print(round(medium['median_income'].mean(),2))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(x='median_income', y='median_house_value', marker='*', s=50, alpha=0.6, data=df)
plt.xlabel('median_income')
plt.ylabel('median_house_value')
plt.title('scatter plot  ')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
df['price_in_thousands']=df['median_house_value']/1000
print(df['price_in_thousands'].head())
```

```python
def price_category(price):
  if price<=100:
    return "budget"
  elif price > 100 and price <= 300:
    return 'mid-range '
  else:
    return 'luxury'
df['category']=df['price_in_thousands'].apply(price_category)
print(df['category'].head())
df['category'].value_counts()
```

```python
round(df['category'].value_counts(normalize=True)*100,2)
```

```python
round(df.groupby('category')['median_income'].mean(),2)
```
